```python
'''In pandas, iloc and loc are two powerful methods for accessing data in a DataFrame. 
They allow you to select and manipulate data based on either integer-location based indexing (iloc) or label-based indexing (loc). 
Understanding how to use these methods is crucial for effective data manipulation in pandas.

iloc: Integer-Location Based Indexing
iloc is used for integer-location based indexing, meaning you select rows and columns by their integer positions. 
It is similar to numpy's array indexing.

loc is used for label-based indexing, meaning you select rows and columns by their labels (names).
It allows for more intuitive data selection when working with labeled data.'''
```

```python
#note 
student_dict={
    'name':['sachin','gupta','shahsi','sandeep'],
    'iq':[100,80,70,50],
    'marks':[80,70,90,60],
    'package':[7,6,4,2]
}
stud=pd.DataFrame(student_dict)
stud.set_index('name',inplace=True)
```

```python
#single row
stud.loc['sachin']
```

```python
#multiple row
stud.loc['sachin':'gupta']
```

```python
#fancy indexing
stud.loc[['gupta','sachin']]
```
