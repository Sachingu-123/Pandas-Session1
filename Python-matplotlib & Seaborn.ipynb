```python
"""
pairplot-->  The pairplot function in Seaborn is used to create a grid of scatter plots (or other plot types) 
showing pairwise relationships between multiple variables in a dataset. 
This is particularly useful for exploring the structure of a dataset and identifying relationships between different variables.
It can also include the distribution of each variable on the diagonal of the grid.
"""
```

```python
'''Pairwise Relationships: Visualizes relationships between multiple variables.
Faceting: Can facet plots by levels of a categorical variable.
Customization: Offers customization options for plot types and aesthetics.'''
```

```python
sns.pairplot(tip)
```

```python
sns.pairplot(tip,hue='sex')
```

```python
''' palette--> refers to a set of colors used for plotting. 
Color palettes help in distinguishing different data series or categories in a plot
, making the visualization more informative and aesthetically pleasing.'''
```

```python
sns.pairplot(tip,hue='sex',palette='pastel')
```

```python
'''A boxplot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based
on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. 
It can also show outliers.
Boxplots are useful for comparing distributions between several groups or datasets.

In Seaborn, boxplots are commonly used to identify and visualize outliers.
Outliers are data points that fall outside the typical range of the dataset 
and are usually represented by individual points beyond the whiskers of the boxplot.
'''
```

```python
sns.boxplot(x='day',y='total_bill',data=tip,palette='rainbow')
```

```python
sns.boxplot(y='day',x='total_bill',data=tip)    #swap the the value of x and y we will get horizontal boxplot
```

```python
'''  The lmplot function in Seaborn is used to create scatter plots with linear regression fits and optionally 
visualize relationships across different subsets of the data using facets.
It is a higher-level interface to sns.regplot and sns.scatterplot, making it easier to create complex visualizations.  '''
```

```python
sns.lmplot(x='total_bill',y='tip',data=tip,aspect=4)     #aspect---is nothing but zoom size
```

```python
'''The catplot function in Seaborn is a categorical plot function that provides
a powerful way to visualize relationships across different subsets of data.
It can be used to create various types of categorical plots depending on the kind of the plot specified.'''
```

```python
sns.catplot(x='day',y='total_bill',data=tip,kind='violin')
```

```python
'''Seaborn does not provide a direct function to create pie charts, but you can still create visually appealing pie charts in
Python by leveraging Seaborn's styling capabilities in conjunction with Matplotlib,
which does have a function for creating pie charts.'''
```
